### To play sound
Useful at the end of long running tasks

```python
# Source: https://stackoverflow.com/questions/17323336/automatically-play-sound-in-ipython-notebook
from IPython.display import Audio
import numpy as np
wave = np.sin(2*np.pi*400*np.arange(10000*2)/10000)
Audio(wave, rate=10000, autoplay=True)
```
